## Podejrzane profile

```python
profiles = pd.read_excel("../../src/profile_ukraina.xlsx",header=1, sheet_name=0)

profiles = profiles[profiles['PLATFORMA']=='Twitter']["LINK PROFILU"]
threat_usernames = []
for profile in profiles:
    try:
        temp_profile = profile.partition("twitter.com/")[2].rstrip('/')
        if len(temp_profile) >= 4:
            threat_usernames.append(temp_profile)
    except:
        pass
threat_usernames
```

## Konta analiza (najaktywniejsze i aktywno≈õƒá kont podejrzanych)

```python
data['username'].value_counts()
```

```python
threat_usernames_occurrences = []
for username in threat_usernames:
    threat_usernames_occurrences.append([len(data[data['username'] == username]), sum(data[data['username'] == username]['nlikes']), sum(data[data['username'] == username]['nreplies']),sum(data[data['username'] == username]['nretweets'])])
```

```python
data_tuples = list(zip(threat_usernames,threat_usernames_occurrences))
pd.DataFrame(data_tuples, columns=['Month','Day']) 
```
